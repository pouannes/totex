```python
def bbox_hex_to_floats(bbox, img_height):
    """
    return the unpacked bbox to be used by fastai bbox.create by converting the hex and
    substracting page height as given by the page element
    """
    bbox = [struct.unpack('!d', bytes.fromhex(t))[0]*(96/72) for t in bbox]
    bbox[1] = img_height - bbox[1]
    bbox[3] = img_height - bbox[3]
    return list(map(round,bbox))
```

```python
def find_pg_size(page):
    page_size = bbox_hex_to_floats(page.attrib['BBox'].split(), 0)[1:3]
    page_size[0] *= -1
    return page_size
```

```python
def find_img_ratios(img, page):
    "return the ratio between the img and what was used as a np array : height, width"
    img_size = list(img.size)
    page_size = find_pg_size(page)
    return np.divide(img_size, page_size)
```

```python
def find_bboxes(elements, img_height, ratios):
    bboxes = []
    for el in elements:
        box = bbox_hex_to_floats(el.attrib['BBox'].split(), img_height)
        box[0], box[1] = box[1] * ratios[0], box[0] * ratios[0]
        box[2], box[3] = box[3] * ratios[1], box[2] * ratios[1]
        bboxes.append(box)
    return bboxes
```

```python
def draw_bbox(img, page, elt_name):
    """
    draw the bbox for the elements with elt_name name (e.g. IsolatedFormula, or */Char), and show the img
    """
    pg_size = find_pg_size(page)
    ratios = find_img_ratios(img, page)
    print(ratios)
    bboxes = bboxes = find_bboxes(root.findall(elt_name), pg_size[0], ratios)
    bbox = ImageBBox.create(bboxes, *img.size)
    img.show(figsize=(40,45), y=bbox)
```

```python
draw_bbox(im, root, "IsolatedFormula")
```

```python
draw_bbox(im, root, "IsolatedFormula/Char")
```

```python
def draw_idx(i):
    img = open_image(images[i])
    xml = images[i].parts[-1].replace('.tif', '.xml')
    tree = ET.parse(path_annotations/xml)
    root = tree.getroot()
    draw_bbox(img, root, "IsolatedFormula")
```
